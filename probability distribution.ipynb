```python

import numpy as np

# Binomial: Number of successful sales out of 100 visits
sales = np.random.binomial(100, 0.30, size=1)


print("Sales: ", sales)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Binomial Distribution Data for Sales
sales = np.random.binomial(100, 0.30, size=1)

# Plotting the data
labels = ['Sales', 'Non-sales']
sizes = [sales[0], 100 - sales[0]]  # Sales and non-sales
colors = ['lightblue', 'lightgreen']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Sales Conversion from 100 Customer Visits')
plt.show()
```

Addition use cases for Binominal

1 -Product Returns: Estimating the number of products returned after a new launch, given a historical return rate, to assess and plan for reverse logistics.
2 -Email Marketing Campaigns: Predicting the number of positive responses (e.g., clicks or purchases) from a given number of promotional emails sent to customers.
3 -Quality Control: Modeling the number of defect-free garments in a sample from a production batch to evaluate the quality control process effectiveness.

```python
# Poisson: Customer complaints per day over a week
complaints = np.random.poisson(2, size=7)


print("Complaints: ", complaints)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Poisson Distribution Data for Customer Complaints
complaints = np.random.poisson(2, size=7)  # 7 days

# Plotting the data
days = range(1, 8)

plt.figure(figsize=(10, 5))
plt.bar(days, complaints, color='salmon')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Complaints')
plt.title('Customer Complaints per Day Over a Week')
plt.xticks(days)
plt.grid(True)
plt.show()
```

Additional Usecases for Poisson

1 - Store Traffic Counts: Predicting the number of customers entering a store per hour or day to optimize staffing and marketing strategies.
2 - Social Media Mentions: Tracking the number of times a brand or product is mentioned on social media platforms per day to gauge brand popularity and customer engagement.
3 - Supply Chain Disruptions: Modeling the number of times supply chain disruptions (like delayed shipments) occur in a month to enhance logistics planning.

```python

# Exponential: Time between customer arrivals, 10 customers
arrival_times = np.random.exponential(scale=12, size=10)

print("Arrival Times: ", arrival_times)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Exponential Distribution Data for Time Between Customer Arrivals
arrival_times = np.random.exponential(scale=12, size=100)  # 100 customers

# Plotting the data
plt.figure(figsize=(10, 6))
sns.histplot(arrival_times, kde=True, color='purple')
plt.title('Histogram of Time Between Customer Arrivals')
plt.xlabel('Time Between Arrivals (minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Additional Usecases for Exponential

1 - Customer Service Wait Times: Estimating the time customers wait on hold when calling a fashion retailer's customer service to improve resource allocation.
2 - Replenishment Cycles: Modeling the time between necessary replenishments of stock for fast-moving items to maintain optimal inventory levels.
3 - Inter-purchase Time: Estimating the time interval between purchases made by a customer on an e-commerce fashion platform to enhance personalized marketing strategies.

```python
# Normal: Daily revenue for a month
daily_revenue = np.random.normal(3000, 500, size=30)

print("Daily Revenue: ", daily_revenue)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Normal Distribution Data for Daily Revenue
daily_revenue = np.random.normal(3000, 500, size=30)  # 30 days

# Plotting the data
plt.figure(figsize=(10, 6))
sns.histplot(daily_revenue, kde=True, color='blue')
plt.title('Histogram of Daily Revenues')
plt.xlabel('Daily Revenue ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Additional Usecases for Normal Distribution

1 - Sales Forecasting: Projecting daily or monthly sales based on historical data to plan for future inventory and staffing needs.
2 - Sizing and Fit Analysis: Analyzing the distribution of garment sizes sold or returned to optimize the range and quantity of sizes offered.
3 - Customer Spend Analysis: Evaluating the distribution of customer spending per visit to tailor pricing, promotions, and loyalty programs effectively.
